Given a string s, find the length of the longest substring without repeating characters.

 
__Example 1__:
```
Input: s = "abcabcbb"
Output: 3
```
Explanation: The answer is "abc", with the length of 3.

__Example 2__:

```
Input: s = "bbbbb"
Output: 1
```
Explanation: The answer is "b", with the length of 1.

__Example 3__:

```
Input: s = "pwwkew"
Output: 3
```
Explanation: The answer is "wke", with the length of 3.
Notice that the answer must be a substring, "pwke" is a subsequence and not a substring.

__Example 4__:

```
Input: s = ""
Output: 0
```

 

Constraints:

    0 <= s.length <= 5 * 104
    s consists of English letters, digits, symbols and spaces.



__Solution 1: Brute force, time limit exceeded__

````

class Solution:
    def lengthOfLongestSubstring(self, s: str) -> int:
      longest = 0 
      n = len(list(s))
      string = ""

      for i in range(n):
        j = i
        while j < n:
          if s[j] in list(string):
            string=""
          string = string + s[j]
          if len(list(string))> longest:
            longest = len(list(string))
          j+=1   
      
      return longest
````

__Solution 2: Sliding Window technique__

````
class Solution:
    def lengthOfLongestSubstring(self, s: str) -> int:
  
      n = len(list(s))
      aux_set = set()
      longest = 0

      j = 0

      for i in range(n):
        while j < n:

          if not (s[j] in aux_set):
            
            aux_set.add(s[j])
            window_size = j-i+1

            if window_size > longest:
              longest = window_size

          else: 
            aux_set.remove(s[i])
            break

          j+=1
      
      return longest
````
