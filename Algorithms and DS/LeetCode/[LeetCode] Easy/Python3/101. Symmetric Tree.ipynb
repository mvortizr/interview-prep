Given a binary tree, check whether it is a mirror of itself (ie, symmetric around its center).

For example, this binary tree [1,2,2,3,4,4,3] is symmetric:

```
    1
   / \
  2   2
 / \ / \
3  4 4  3

```

But the following [1,2,2,null,3,null,3] is not:

```
    1
   / \
  2   2
   \   \
   3    3
```
 

````
class Solution:
    
    def isSymmetric(self, root: TreeNode) -> bool:
        if root != None:
            return self.check_mirror(root.right,root.left)
        return True
    
    def check_mirror(self,node_right:TreeNode,node_left:TreeNode)->bool:      
        
        if node_right == None and node_left == None:
            return True
        
        if node_right == None or node_left == None:
            return False
        
        if node_right.val != node_left.val:
            return False
        
        return self.check_mirror(node_right.right,node_left.left) and self.check_mirror(node_left.right, node_right.left)
        

 
````
