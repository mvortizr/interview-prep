# __Reverse Integer__

Given a 32-bit signed integer, reverse digits of an integer.

Example 1:
```
Input: 123
Output: 321
```
Example 2:

```
Input: -123
Output: -321
```
Example 3:
```
Input: 120
Output: 21
```
Example 4:
```
Input: 1534236469 
Output: 0
```
Note:
Assume we are dealing with an environment which could only store integers within the 32-bit signed integer range: [−2^31,  2^31 − 1]. For the purpose of this problem, assume that your function returns 0 when the reversed integer overflows.


### __My solution__

We can extract the last digit for a number by dividing by 10 `popped_digit = number/10` and we build the result while iterating multiplying `*10` and adding the popped digit. 

We have to be careful that the result doesn't overflows the system while we are building it. This check has to be done before creating the overflowing condition, because some systems when they overflow just puts random values to the number. 

Python is kind of messy when doing operations with negative values, so my option here was to use the abs function. Another thing that can be done is to to create special divide and mod functions such as [this video](https://www.youtube.com/watch?v=Io9ujnnR4sI) but I consider that to be way overkill.

````
class Solution:


  def number_overflows(self,sign:int,digit:int,result:int)->int:
    #This function check if the result will overflow the system
    MAX_POS = 2 ** 31 -1 #2147483647 for positive numbers
    MAX_NEG = 2 ** 31 #-2147483648 for negative numbers

    #For positive numbers
    if sign == 1 and (result > MAX_POS//10 or (result == MAX_POS//10 and digit>7) ):
      return 1
    
    #For negative numbers
    if sign == -1 and (result > MAX_NEG//10 or (result == MAX_NEG and digit>8) ):
      return 1

    return 0 

  

  def reverse(self,x:int)-> int:  
    
    result = 0

    #Get the sign
    sign = -1 if x<0 else 1  # 1 positive, -1 negative
    x = abs(x)

    #Loop to rebuild x
    while x > 0:
      digit = x % 10 
      if self.number_overflows(sign,digit,result): # Return if overflows
        return 0    
      result = result*10 + digit   
      x = x // 10
    return result * sign
    
````

````
sol = Solution()
sol.reverse(1534236469)
````
