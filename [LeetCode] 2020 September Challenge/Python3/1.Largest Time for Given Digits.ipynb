Given an array of 4 digits, return the largest 24 hour time that can be made.

The smallest 24 hour time is 00:00, and the largest is 23:59.  Starting from 00:00, a time is larger if more time has elapsed since midnight.

Return the answer as a string of length 5.  If no valid time can be made, return an empty string.


Example 1:

```
Input: [1,2,3,4]
Output: "23:41"
```

Example 2:
```
Input: [5,5,5,5]
Output: ""
```


````
class Solution:
      
    def getDigit(self,A,max)->int: 
      print('A', A)     
      for i in range(max,-1,-1):  
          #print('A',A)  
          if i in A:
              A.remove(i)
              print('A dentro', A)
              return i  
      return -1

    def getSecondDigit(self,A,max,first)->int:
        if first<2:
          return self.getDigit(A,max)
        else:
          return self.getDigit(A,3)
        
    def largestTimeFromDigits(self, A) -> str:     



        
            
        #First digit - hours, range [0,2]     
        firstDigit = self.getDigit(A,2)
        print('uno')

        #Second digit - hours, range [0,3]
        secondDigit = self.getSecondDigit(A,9,firstDigit)
        print('dos')

        #Third digit - minutes, range[0,5]
        thirdDigit = self.getDigit(A,5)
        print('tres')

        #Fourth digit - minutes, range[0,9]
        fourthDigit = self.getDigit(A,9)
        print('cuatro')
        
        if firstDigit != -1 and secondDigit != -1 and thirdDigit!=-1 and fourthDigit !=-1: 
          return f'{firstDigit}{secondDigit}:{thirdDigit}{fourthDigit}'
        else:
          print(f'{firstDigit}{secondDigit}:{thirdDigit}{fourthDigit}')
          return ''

solution = Solution()
````
